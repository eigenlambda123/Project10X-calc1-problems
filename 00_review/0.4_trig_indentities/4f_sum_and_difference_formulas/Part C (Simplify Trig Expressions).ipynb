### **Problems**

**11.** Simplify: $\sin(45^\circ + 30^\circ)$  
**12.** Simplify: $\cos(60^\circ - 45^\circ)$  
**13.** Simplify: $\tan(45^\circ - 30^\circ)$


---

### **Solutions**


**11.** Use the identity:  
$$
\sin(a + b) = \sin(a)\cos(b) + \cos(a)\sin(b)
$$

Substitute values:
- $\sin(45^\circ) = \dfrac{\sqrt{2}}{2}$  
- $\cos(30^\circ) = \dfrac{\sqrt{3}}{2}$  
- $\cos(45^\circ) = \dfrac{\sqrt{2}}{2}$  
- $\sin(30^\circ) = \dfrac{1}{2}$

So:
$$
\sin(75^\circ) = \dfrac{\sqrt{2}}{2} \cdot \dfrac{\sqrt{3}}{2} + \dfrac{\sqrt{2}}{2} \cdot \dfrac{1}{2}
= \dfrac{\sqrt{6} + \sqrt{2}}{4}
$$

**Answer:**  
$\boxed{\dfrac{\sqrt{6} + \sqrt{2}}{4}}$


**12.** Substitute values:
- $\cos(60^\circ) = \dfrac{1}{2}$  
- $\cos(45^\circ) = \dfrac{\sqrt{2}}{2}$  
- $\sin(60^\circ) = \dfrac{\sqrt{3}}{2}$  
- $\sin(45^\circ) = \dfrac{\sqrt{2}}{2}$

So:
$$
= \dfrac{1}{2} \cdot \dfrac{\sqrt{2}}{2} + \dfrac{\sqrt{3}}{2} \cdot \dfrac{\sqrt{2}}{2}
= \dfrac{\sqrt{2}}{4} + \dfrac{\sqrt{6}}{4}
= \dfrac{\sqrt{2} + \sqrt{6}}{4}
$$

**Answer:**  
$\boxed{\dfrac{\sqrt{2} + \sqrt{6}}{4}}$



**13.** Use:
$$
\tan(a - b) = \dfrac{\tan(a) - \tan(b)}{1 + \tan(a)\tan(b)}
$$

Substitute:
- $\tan(45^\circ) = 1$  
- $\tan(30^\circ) = \dfrac{1}{\sqrt{3}}$

So:
$$
\tan(15^\circ) = \dfrac{1 - \dfrac{1}{\sqrt{3}}}{1 + 1 \cdot \dfrac{1}{\sqrt{3}}}
= \dfrac{\dfrac{\sqrt{3} - 1}{\sqrt{3}}}{\dfrac{\sqrt{3} + 1}{\sqrt{3}}}
= \dfrac{\sqrt{3} - 1}{\sqrt{3} + 1}
$$

Rationalize:
$$
= \dfrac{\sqrt{3} - 1}{\sqrt{3} + 1} \cdot \dfrac{\sqrt{3} - 1}{\sqrt{3} - 1}
= \dfrac{(\sqrt{3} - 1)^2}{(\sqrt{3} + 1)(\sqrt{3} - 1)}
= \dfrac{3 - 2\sqrt{3} + 1}{3 - 1}
= \dfrac{4 - 2\sqrt{3}}{2} = 2 - \sqrt{3}
$$

**Answer:**  
$\boxed{2 - \sqrt{3}}$
