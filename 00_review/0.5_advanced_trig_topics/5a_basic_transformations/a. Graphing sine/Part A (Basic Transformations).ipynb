### **Problems**

**1.** Graph: $y = \sin(x)$  
*(Base graph. Label all key points over $[0,\ 2\pi]$)*

**2.** Graph: $y = 2\sin(x)$  
*(Amplitude change only)*

**3.** Graph: $y = \sin(2x)$  
*(Period change only)*

**4.** Graph: $y = \sin\left(x - \dfrac{\pi}{2}\right)$  
*(Phase shift to the right)*

**5.** Graph: $y = \sin(x) + 1$  
*(Vertical shift upward)*

---
  
### **Solutions**

**1.** $y = \sin(x)$  
- **Amplitude**: 1  
- **Period**: $2\pi$  
- **Phase Shift**: 0  
- **Vertical Shift**: 0  
- **Key Points** (one full cycle):
  - $(0,\ 0)$
  - $\left(\dfrac{\pi}{2},\ 1\right)$
  - $(\pi,\ 0)$
  - $\left(\dfrac{3\pi}{2},\ -1\right)$
  - $(2\pi,\ 0)$


**2.** $y = 2\sin(x)$  
- **Amplitude**: 2  
- **Period**: $2\pi$  
- **Phase Shift**: 0  
- **Vertical Shift**: 0  
- **Key Points**:
  - $(0,\ 0)$
  - $\left(\dfrac{\pi}{2},\ 2\right)$
  - $(\pi,\ 0)$
  - $\left(\dfrac{3\pi}{2},\ -2\right)$
  - $(2\pi,\ 0)$


**3.** $y = \sin(2x)$  
- **Amplitude**: 1  
- **Period**: $\dfrac{2\pi}{2} = \pi$  
- **Phase Shift**: 0  
- **Vertical Shift**: 0  
- **Key Points** (one cycle over $[0, \pi]$):
  - $(0,\ 0)$
  - $\left(\dfrac{\pi}{4},\ 1\right)$
  - $\left(\dfrac{\pi}{2},\ 0\right)$
  - $\left(\dfrac{3\pi}{4},\ -1\right)$
  - $(\pi,\ 0)$


**4.** $y = \sin\left(x - \dfrac{\pi}{2}\right)$  
- **Amplitude**: 1  
- **Period**: $2\pi$  
- **Phase Shift**: $\dfrac{\pi}{2}$ to the **right**  
- **Vertical Shift**: 0  
- **Key Points** (shift original sine graph right by $\dfrac{\pi}{2}$):
  - $\left(\dfrac{\pi}{2},\ 0\right)$
  - $(\pi,\ 1)$
  - $\left(\dfrac{3\pi}{2},\ 0\right)$
  - $(2\pi,\ -1)$
  - $\left(\dfrac{5\pi}{2},\ 0\right)$


**5.** $y = \sin(x) + 1$  
- **Amplitude**: 1  
- **Period**: $2\pi$  
- **Phase Shift**: 0  
- **Vertical Shift**: **+1**  
- **Range**: $[0, 2]$  
- **Key Points** (shift all sine values up by 1):
  - $(0,\ 1)$
  - $\left(\dfrac{\pi}{2},\ 2\right)$
  - $(\pi,\ 1)$
  - $\left(\dfrac{3\pi}{2},\ 0\right)$
  - $(2\pi,\ 1)$

### **Graphs**

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 2 * np.pi, 1000)

fig, axes = plt.subplots(5, 1, figsize=(8, 10), sharex=True)
fig.suptitle("Graphing Sine — Part A", fontsize=14)

# 1. y = sin(x)
axes[0].plot(x, np.sin(x), color='b')
axes[0].set_title(r"$y = \sin(x)$")
axes[0].set_ylim(-2.5, 2.5)

# 2. y = 2sin(x)
axes[1].plot(x, 2 * np.sin(x), color='g')
axes[1].set_title(r"$y = 2\sin(x)$")
axes[1].set_ylim(-2.5, 2.5)

# 3. y = sin(2x)
axes[2].plot(x, np.sin(2 * x), color='r')
axes[2].set_title(r"$y = \sin(2x)$")
axes[2].set_ylim(-2.5, 2.5)

# 4. y = sin(x - pi/2)
axes[3].plot(x, np.sin(x - np.pi/2), color='orange')
axes[3].set_title(r"$y = \sin\left(x - \frac{\pi}{2}\right)$")
axes[3].set_ylim(-2.5, 2.5)

# 5. y = sin(x) + 1
axes[4].plot(x, np.sin(x) + 1, color='purple')
axes[4].set_title(r"$y = \sin(x) + 1$")
axes[4].set_ylim(-2.5, 2.5)

# Formatting
for ax in axes:
    ax.axhline(0, color='black', linewidth=0.5)
    ax.set_xlim(0, 2 * np.pi)
    ax.label_outer()  # Hide x-labels except bottom

# X-ticks as multiples of pi
xticks = [0, np.pi/2, np.pi, 3*np.pi/2, 2*np.pi]
xtick_labels = [r"$0$", r"$\frac{\pi}{2}$", r"$\pi$", r"$\frac{3\pi}{2}$", r"$2\pi$"]
axes[-1].set_xticks(xticks)
axes[-1].set_xticklabels(xtick_labels)

plt.tight_layout(rect=[0, 0, 1, 0.97])
plt.show()
```
