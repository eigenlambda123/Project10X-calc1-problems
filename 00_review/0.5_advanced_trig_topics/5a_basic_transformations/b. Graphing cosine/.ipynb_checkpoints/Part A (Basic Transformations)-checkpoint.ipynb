### **Problems**

**1.** Graph: $y = \sin(x)$  
*(Base graph. Label all key points over $[0,\ 2\pi]$)*

**2.** Graph: $y = 2\sin(x)$  
*(Amplitude change only)*

**3.** Graph: $y = \sin(2x)$  
*(Period change only)*

**4.** Graph: $y = \sin\left(x - \dfrac{\pi}{2}\right)$  
*(Phase shift to the right)*

**5.** Graph: $y = \sin(x) + 1$  
*(Vertical shift upward)*

---
### **Solutions**

**1.** $y = \cos(x)$  
- **Amplitude**: 1  
- **Period**: $2\pi$  
- **Phase Shift**: 0  
- **Vertical Shift**: 0  
- **Key Points** (one full cycle):
  - $(0,\ 1)$  
  - $\left(\dfrac{\pi}{2},\ 0\right)$  
  - $(\pi,\ -1)$  
  - $\left(\dfrac{3\pi}{2},\ 0\right)$  
  - $(2\pi,\ 1)$


**2.** $y = 2\cos(x)$  
- **Amplitude**: 2  
- **Period**: $2\pi$  
- **Phase Shift**: 0  
- **Vertical Shift**: 0  
- **Key Points**:
  - $(0,\ 2)$  
  - $\left(\dfrac{\pi}{2},\ 0\right)$  
  - $(\pi,\ -2)$  
  - $\left(\dfrac{3\pi}{2},\ 0\right)$  
  - $(2\pi,\ 2)$


**3.** $y = \cos(3x)$  
- **Amplitude**: 1  
- **Period**: $\dfrac{2\pi}{3}$  
- **Phase Shift**: 0  
- **Vertical Shift**: 0  
- **Key Points** (one cycle over $[0,\ \dfrac{2\pi}{3}]$):
  - $(0,\ 1)$  
  - $\left(\dfrac{\pi}{6},\ 0\right)$  
  - $\left(\dfrac{\pi}{3},\ -1\right)$  
  - $\left(\dfrac{\pi}{2},\ 0\right)$  
  - $\left(\dfrac{2\pi}{3},\ 1\right)$


**4.** $y = \cos\left(x - \dfrac{\pi}{3}\right)$  
- **Amplitude**: 1  
- **Period**: $2\pi$  
- **Phase Shift**: $\dfrac{\pi}{3}$ to the **right**  
- **Vertical Shift**: 0  
- **Key Points**:
  - $\left(\dfrac{\pi}{3},\ 1\right)$  
  - $\left(\dfrac{5\pi}{6},\ 0\right)$  
  - $\left(\dfrac{4\pi}{3},\ -1\right)$  
  - $\left(\dfrac{11\pi}{6},\ 0\right)$  
  - $\left(\dfrac{7\pi}{3},\ 1\right)$


**5.** $y = \cos(x) - 1$  
- **Amplitude**: 1  
- **Period**: $2\pi$  
- **Phase Shift**: 0  
- **Vertical Shift**: $-1$  
- **Range**: $[-2,\ 0]$  
- **Key Points**:
  - $(0,\ 0)$  
  - $\left(\dfrac{\pi}{2},\ -1\right)$  
  - $(\pi,\ -2)$  
  - $\left(\dfrac{3\pi}{2},\ -1\right)$  
  - $(2\pi,\ 0)$


---
### **Graphs**

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 2 * np.pi, 1000)

fig, axes = plt.subplots(5, 1, figsize=(8, 10), sharex=True)
fig.suptitle("Graphing Cosine – Part A", fontsize=14)

# 1. y = cos(x)
axes[0].plot(x, np.cos(x), color='b')
axes[0].set_title(r"$y = \cos(x)$")
axes[0].set_ylim(-2.5, 2.5)

# 2. y = 2cos(x)
axes[1].plot(x, 2 * np.cos(x), color='g')
axes[1].set_title(r"$y = 2\cos(x)$")
axes[1].set_ylim(-2.5, 2.5)

# 3. y = cos(3x)
axes[2].plot(x, np.cos(3 * x), color='r')
axes[2].set_title(r"$y = \cos(3x)$")
axes[2].set_ylim(-2.5, 2.5)

# 4. y = cos(x - pi/3)
axes[3].plot(x, np.cos(x - np.pi / 3), color='orange')
axes[3].set_title(r"$y = \cos\left(x - \frac{\pi}{3}\right)$")
axes[3].set_ylim(-2.5, 2.5)

# 5. y = cos(x) - 1
axes[4].plot(x, np.cos(x) - 1, color='purple')
axes[4].set_title(r"$y = \cos(x) - 1$")
axes[4].set_ylim(-2.5, 2.5)

# Format axes
for ax in axes:
    ax.axhline(0, color='black', linewidth=0.5)
    ax.set_xlim(0, 2 * np.pi)
    ax.label_outer()

# π-based x-ticks
xticks = [0, np.pi/2, np.pi, 3*np.pi/2, 2*np.pi]
xtick_labels = [r"$0$", r"$\frac{\pi}{2}$", r"$\pi$", r"$\frac{3\pi}{2}$", r"$2\pi$"]
axes[-1].set_xticks(xticks)
axes[-1].set_xticklabels(xtick_labels)

plt.tight_layout(rect=[0, 0, 1, 0.97])
plt.show()
```
