---
### **Solutions**

**6.** $y = 3\cos(2x)$  
- **Amplitude**: 3  
- **Period**: $\dfrac{2\pi}{2} = \pi$  
- **Phase Shift**: 0  
- **Vertical Shift**: 0  
- **Key Points** (one cycle over $[0, \pi]$):
  - $(0,\ 3)$  
  - $\left(\dfrac{\pi}{4},\ 0\right)$  
  - $\left(\dfrac{\pi}{2},\ -3\right)$  
  - $\left(\dfrac{3\pi}{4},\ 0\right)$  
  - $(\pi,\ 3)$  


**7.** $y = -\cos\left(x + \dfrac{\pi}{4}\right)$  
- **Amplitude**: 1 (reflected)  
- **Period**: $2\pi$  
- **Phase Shift**: $\dfrac{\pi}{4}$ **left**  
- **Vertical Shift**: 0  
- **Key Points**:
  - $\left(-\dfrac{\pi}{4},\ -1\right)$  
  - $(0,\ 0)$  
  - $\left(\dfrac{\pi}{4},\ 1\right)$  
  - $\left(\dfrac{\pi}{2},\ 0\right)$  
  - $\left(\dfrac{3\pi}{4},\ -1\right)$  


**8.** $y = 0.5\cos(4x - \pi)$  
- **Amplitude**: $0.5$  
- **Period**: $\dfrac{2\pi}{4} = \dfrac{\pi}{2}$  
- **Phase Shift**: Solve $4x - \pi = 0 \Rightarrow x = \dfrac{\pi}{4}$ → **Right**  
- **Vertical Shift**: 0  
- **Key Points** (one cycle starting at $\dfrac{\pi}{4}$):
  - $\left(\dfrac{\pi}{4},\ 0.5\right)$  
  - $\left(\dfrac{3\pi}{8},\ 0\right)$  
  - $\left(\dfrac{\pi}{2},\ -0.5\right)$  
  - $\left(\dfrac{5\pi}{8},\ 0\right)$  
  - $\left(\dfrac{3\pi}{4},\ 0.5\right)$  


**9.** $y = 2\cos\left(\dfrac{1}{2}x + \dfrac{\pi}{6}\right) + 2$  
- **Amplitude**: 2  
- **Period**: $\dfrac{2\pi}{1/2} = 4\pi$  
- **Phase Shift**: Solve $\dfrac{1}{2}x + \dfrac{\pi}{6} = 0 \Rightarrow x = -\dfrac{\pi}{3}$ → **Left**  
- **Vertical Shift**: $+2$  
- **Key Points** (one cycle starts at $x = -\dfrac{\pi}{3}$ and ends at $x = \dfrac{11\pi}{3}$):
  - $(-\dfrac{\pi}{3},\ 4)$  
  - $(\dfrac{5\pi}{6},\ 2)$  
  - $(\dfrac{8\pi}{3},\ 0)$  
  - $(\dfrac{17\pi}{6},\ 2)$  
  - $(\dfrac{11\pi}{3},\ 4)$

### **Graphs**

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-2 * np.pi, 4 * np.pi, 1000)

fig, axes = plt.subplots(4, 1, figsize=(10, 10), sharex=True)
fig.suptitle("Graphing Cosine – Part B", fontsize=15)

# 6. y = 3cos(2x)
axes[0].plot(x, 3 * np.cos(2 * x), color='b')
axes[0].set_title(r"$y = 3\cos(2x)$")
axes[0].set_ylim(-4, 4)

# 7. y = -cos(x + pi/4)
axes[1].plot(x, -np.cos(x + np.pi / 4), color='g')
axes[1].set_title(r"$y = -\cos\left(x + \frac{\pi}{4}\right)$")
axes[1].set_ylim(-1.5, 1.5)

# 8. y = 0.5cos(4x - pi)
axes[2].plot(x, 0.5 * np.cos(4 * x - np.pi), color='r')
axes[2].set_title(r"$y = 0.5\cos\left(4x - \pi\right)$")
axes[2].set_ylim(-1, 1)

# 9. y = 2cos(0.5x + pi/6) + 2
axes[3].plot(x, 2 * np.cos(0.5 * x + np.pi / 6) + 2, color='orange')
axes[3].set_title(r"$y = 2\cos\left(\frac{1}{2}x + \frac{\pi}{6}\right) + 2$")
axes[3].set_ylim(-1, 5)

# Common formatting
for ax in axes:
    ax.axhline(0, color='black', linewidth=0.5)
    ax.set_xlim(-np.pi, 3 * np.pi)
    ax.label_outer()

# π-based ticks
xticks = [-np.pi, 0, np.pi, 2*np.pi, 3*np.pi]
xtick_labels = [r"$-\pi$", r"$0$", r"$\pi$", r"$2\pi$", r"$3\pi$"]
axes[-1].set_xticks(xticks)
axes[-1].set_xticklabels(xtick_labels)

plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.show()
```
