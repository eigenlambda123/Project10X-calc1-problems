### **Problems**
Sketch and analyze the following tangent functions. Find:
- Period
- Phase shift
- Equations of vertical asymptotes
- 3 key points (left, center, right of one period)

**1.** $y = \tan(x)$  
**2.** $y = \tan(2x)$  
**3.** $y = \tan\left(x - \dfrac{\pi}{4}\right)$  
**4.** $y = -\tan(x)$  
**5.** $y = \tan(x) + 1$  

---
### **Solutions**

**1.** $y = \tan(x)$  
- **Amplitude**: *Not defined*  
- **Period**: $\pi$  
- **Phase Shift**: $0$  
- **Vertical Shift**: $0$  
- **Vertical Asymptotes**: $x = -\dfrac{\pi}{2},\ \dfrac{\pi}{2}$  
- **Key Points**:
  - $\left(-\dfrac{\pi}{4},\ -1\right)$  
  - $(0,\ 0)$  
  - $\left(\dfrac{\pi}{4},\ 1\right)$  
  

**2.** $y = \tan(2x)$  
- **Amplitude**: *Not defined*  
- **Period**: $\dfrac{\pi}{2}$  
- **Phase Shift**: $0$  
- **Vertical Shift**: $0$  
- **Vertical Asymptotes**: $x = -\dfrac{\pi}{4},\ \dfrac{\pi}{4}$  
- **Key Points**:
  - $\left(-\dfrac{\pi}{8},\ -1\right)$  
  - $(0,\ 0)$  
  - $\left(\dfrac{\pi}{8},\ 1\right)$  


**3.** $y = \tan\left(x - \dfrac{\pi}{4}\right)$  
- **Amplitude**: *Not defined*  
- **Period**: $\pi$  
- **Phase Shift**: $\dfrac{\pi}{4}$ to the **right**  
- **Vertical Shift**: $0$  
- **Vertical Asymptotes**:  
  $x = -\dfrac{\pi}{4} + \dfrac{\pi}{2} = \dfrac{\pi}{4}$  
  and  
  $x = -\dfrac{\pi}{4} - \dfrac{\pi}{2} = -\dfrac{3\pi}{4}$  
- **Key Points**:
  - $\left(0,\ -1\right)$  
  - $\left(\dfrac{\pi}{4},\ 0\right)$  
  - $\left(\dfrac{\pi}{2},\ 1\right)$  
  

**4.** $y = -\tan(x)$  
- **Amplitude**: *Not defined*  
- **Period**: $\pi$  
- **Phase Shift**: $0$  
- **Vertical Shift**: $0$  
- **Vertical Asymptotes**: $x = -\dfrac{\pi}{2},\ \dfrac{\pi}{2}$  
- **Key Points**:
  - $\left(-\dfrac{\pi}{4},\ 1\right)$  
  - $(0,\ 0)$  
  - $\left(\dfrac{\pi}{4},\ -1\right)$  

> Reflected across the x-axis.


**5.** $y = \tan(x) + 1$  
- **Amplitude**: *Not defined*  
- **Period**: $\pi$  
- **Phase Shift**: $0$  
- **Vertical Shift**: **+1**  
- **Vertical Asymptotes**: $x = -\dfrac{\pi}{2},\ \dfrac{\pi}{2}$  
- **Key Points**:
  - $\left(-\dfrac{\pi}{4},\ 0\right)$  
  - $(0,\ 1)$  
  - $\left(\dfrac{\pi}{4},\ 2\right)$  
  
> The entire graph is shifted **up** by 1 unit.


### **Graphs**

```python
import numpy as np
import matplotlib.pyplot as plt

# Set up domain (avoid discontinuities by clipping spikes)
x = np.linspace(-np.pi, np.pi, 1000)

# Helper to clip extreme values (vertical asymptote zones)
def safe_tan(y):
    y = y.copy()
    y[np.abs(y) > 10] = np.nan
    return y

fig, axes = plt.subplots(5, 1, figsize=(8, 12), sharex=True)
fig.suptitle("Graphing Tangent – Part A", fontsize=14)

# 1. y = tan(x)
axes[0].plot(x, safe_tan(np.tan(x)), color='b')
axes[0].set_title(r"$y = \tan(x)$")

# 2. y = tan(2x)
axes[1].plot(x, safe_tan(np.tan(2 * x)), color='g')
axes[1].set_title(r"$y = \tan(2x)$")

# 3. y = tan(x - π/4)
axes[2].plot(x, safe_tan(np.tan(x - np.pi / 4)), color='r')
axes[2].set_title(r"$y = \tan\left(x - \frac{\pi}{4}\right)$")

# 4. y = -tan(x)
axes[3].plot(x, safe_tan(-np.tan(x)), color='orange')
axes[3].set_title(r"$y = -\tan(x)$")

# 5. y = tan(x) + 1
axes[4].plot(x, safe_tan(np.tan(x) + 1), color='purple')
axes[4].set_title(r"$y = \tan(x) + 1$")

# Apply formatting to all subplots
for ax in axes:
    ax.axhline(0, color='black', linewidth=0.5)
    ax.set_ylim(-5, 5)
    ax.set_xlim(-np.pi, np.pi)
    ax.label_outer()

    # Draw vertical asymptotes for tan(x) at x = ±π/2, ±3π/2, ...
    for n in range(-2, 3):
        ax.axvline((2 * n + 1) * np.pi / 2, color='gray', linestyle='--', alpha=0.4)

# π-based x-ticks
xticks = [-np.pi, -np.pi/2, 0, np.pi/2, np.pi]
xtick_labels = [r"$-\pi$", r"$-\frac{\pi}{2}$", r"$0$", r"$\frac{\pi}{2}$", r"$\pi$"]
axes[-1].set_xticks(xticks)
axes[-1].set_xticklabels(xtick_labels)

plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.show()
```
