### **Problems**

State the **domain and range** of each of the following functions:

1. $f(x) = \sin^{-1}(x)$
2. $g(x) = \cos^{-1}(x)$
3. $h(x) = \tan^{-1}(x)$

Also identify which of these are **even, odd, or neither**.

---
### **Solutions**

**1.** $\sin^{-1}\left(\frac{\sqrt{3}}{2}\right)$

We are looking for the angle $\theta$ such that:

$$
\sin(\theta) = \frac{\sqrt{3}}{2}
$$

within the principal range of $\sin^{-1}(x)$:

$$
\left[-\frac{\pi}{2}, \frac{\pi}{2}\right]
$$

So,

$$
\sin^{-1}\left(\frac{\sqrt{3}}{2}\right) = \boxed{\frac{\pi}{3}}
$$


**2.** $\cos^{-1}\left(-\frac{1}{2}\right)$

We are looking for $\theta$ such that:

$$
\cos(\theta) = -\frac{1}{2}
$$

within the range of $\cos^{-1}(x)$:

$$
[0, \pi]
$$

So,

$$
\cos^{-1}\left(-\frac{1}{2}\right) = \boxed{\frac{2\pi}{3}}
$$


**3.** $\tan^{-1}(\sqrt{3})$

We are looking for $\theta$ such that:

$$
\tan(\theta) = \sqrt{3}
$$

within the range of $\tan^{-1}(x)$:

$$
\left(-\frac{\pi}{2}, \frac{\pi}{2}\right)
$$

So,

$$
\tan^{-1}(\sqrt{3}) = \boxed{\frac{\pi}{3}}
$$
