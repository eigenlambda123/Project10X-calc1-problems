### **Problems**

Prove the following identities:

1. $\tan^{-1}(x) + \tan^{-1}(y) = \tan^{-1}\left(\frac{x + y}{1 - xy}\right)$ (for $xy < 1$)
2. $\sin^{-1}(x) + \cos^{-1}(x) = \frac{\pi}{2}$
3. $\tan^{-1}(1/x) = \frac{\pi}{2} - \tan^{-1}(x)$ (for $x > 0$)
4. $\cos(\tan^{-1}(x)) = \frac{1}{\sqrt{1 + x^2}}$

---
### **Solutions**

**1.** Let:
$$
\alpha = \tan^{-1}(x), \quad \beta = \tan^{-1}(y)
$$

Then:
$$
\tan(\alpha) = x, \quad \tan(\beta) = y
$$

By the tangent addition formula:
$$
\tan(\alpha + \beta) = \frac{\tan(\alpha) + \tan(\beta)}{1 - \tan(\alpha)\tan(\beta)} = \frac{x + y}{1 - xy}
$$

So:
$$
\alpha + \beta = \tan^{-1}\left(\frac{x + y}{1 - xy}\right)
$$

**Final Result:**
$$
\boxed{\tan^{-1}(x) + \tan^{-1}(y) = \tan^{-1}\left(\frac{x + y}{1 - xy}\right)}
$$

**2.** Let:
$$
\theta = \sin^{-1}(x)
$$

Then:
$$
\sin(\theta) = x \quad \Rightarrow \quad \cos(\theta) = \sqrt{1 - x^2}
$$

So:
$$
\cos^{-1}(\cos(\theta)) = \cos^{-1}(\sqrt{1 - x^2}) = \frac{\pi}{2} - \theta
$$

Hence:
$$
\sin^{-1}(x) + \cos^{-1}(x) = \theta + \left(\frac{\pi}{2} - \theta\right) = \frac{\pi}{2}
$$

**Final Result:**
$$
\boxed{\sin^{-1}(x) + \cos^{-1}(x) = \frac{\pi}{2}}
$$


**3.** Let:
$$
\theta = \tan^{-1}(x) \quad \Rightarrow \quad \tan(\theta) = x
$$

Then:
$$
\cot(\theta) = \frac{1}{x} \quad \Rightarrow \quad \tan^{-1}\left(\frac{1}{x}\right) = \cot^{-1}(x)
$$

And:
$$
\cot^{-1}(x) = \frac{\pi}{2} - \tan^{-1}(x)
$$

**Final Result:**
$$
\boxed{\tan^{-1}\left(\frac{1}{x}\right) = \frac{\pi}{2} - \tan^{-1}(x)}
$$


**4.** Let:
$$
\theta = \tan^{-1}(x) \Rightarrow \tan(\theta) = x = \frac{\text{opposite}}{\text{adjacent}} = \frac{x}{1}
$$

From a right triangle:
- Opposite = $x$
- Adjacent = $1$
- Hypotenuse = $\sqrt{1 + x^2}$

Then:
$$
\cos(\theta) = \frac{\text{adjacent}}{\text{hypotenuse}} = \frac{1}{\sqrt{1 + x^2}}
$$

**Final Result:**
$$
\boxed{\cos(\tan^{-1}(x)) = \frac{1}{\sqrt{1 + x^2}}}
$$
