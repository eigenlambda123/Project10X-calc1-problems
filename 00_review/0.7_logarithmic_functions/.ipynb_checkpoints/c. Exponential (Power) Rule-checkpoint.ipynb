### **Problems**

**Problem 1**
Simplify:
$$
\log_2(3^4)
$$


**Problem 2**
Rewrite using the power rule:
$$
\log_b(x^5)
$$


**Problem 3**
Simplify:
$$
\ln(e^7)
$$


**Problem 4**
Rewrite the expression using log properties:
$$
\log_{10}(100^x)
$$


**Problem 5**
Simplify and evaluate:
$$
\log_3(81^2)
$$


**Problem 6**
Given: $\log_7 x = 4$, find:
$$
\log_7(x^3)
$$


**Problem 7**
Rewrite the expression using the power rule:
$$
\ln(\sqrt{x})
$$


**Problem 8**
Simplify fully:
$$
\log_5((2x)^3)
$$


**Problem 9**
True or False?
$$
\log_b(x^2y^3) = 2\log_b x + 3\log_b y
$$


**Problem 10**
Rewrite in expanded form:
$$
\ln(a^2b^5)
$$

---

### **Solutions**

**1.** $\log_2(3^4) = 4 \log_2 3 = \boxed{4 \log_2 3}$

**2.** $\log_b(x^5) = 5 \log_b x = \boxed{5 \log_b x}$

**3.** $\ln(e^7) = 7 \ln e = 7(1) = \boxed{7}$

**4.** $\log_{10}(100^x) = x \log_{10} 100 = x \cdot 2 = \boxed{2x}$

**5.** $\log_3(81^2) = 2 \log_3 81 = 2 \cdot 4 = \boxed{8}$

**6.** $\log_7(x^3) = 3 \log_7 x = 3(4) = \boxed{12}$

**7.** $\ln(\sqrt{x}) = \ln(x^{1/2}) = \frac{1}{2} \ln x = \boxed{\frac{1}{2} \ln x}$

**8.** $\log_5((2x)^3) = \log_5(8x^3) = \log_5 8 + \log_5 x^3 = \log_5 8 + 3 \log_5 x = \boxed{\log_5 8 + 3 \log_5 x}$

**9.** $\log_b(x^2y^3) = 2 \log_b x + 3 \log_b y$ â€” **True**

**10.** $\ln(a^2b^5) = \ln a^2 + \ln b^5 = 2 \ln a + 5 \ln b = \boxed{2 \ln a + 5 \ln b}$
