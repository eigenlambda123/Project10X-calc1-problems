### **Problems**

**For each of the following, sketch the line and find the slope:**

1. A line that goes through $(0,0)$ and $(2,4)$
2. A line parallel to $y = 2x - 3$ that goes through $(1, 1)$
3. A horizontal line through the point $(5, -2)$
4. A vertical line through the point $(-1, 3)$
5. A line that decreases 3 units vertically for every 2 units it goes right

---

### **Solutions**

```python
import matplotlib.pyplot as plt
import numpy as np

fig, axs = plt.subplots(3, 2, figsize=(12, 12))
axs = axs.flatten()

# 1. Through (0,0) and (2,4)
x1 = np.array([0, 2])
y1 = np.array([0, 4])
axs[0].plot(x1, y1, label='Slope = 2', color='blue')
axs[0].scatter(x1, y1, color='black')
axs[0].set_title("Line through (0,0) and (2,4)")

# 2. Parallel to y = 2x - 3 through (1,1)
x2 = np.linspace(-2, 4, 100)
y2 = 2 * x2 - 1  # same slope = 2, passes through (1,1)
axs[1].plot(x2, y2, label='Slope = 2', color='green')
axs[1].scatter([1], [1], color='black')
axs[1].set_title("Line parallel to y = 2x - 3 through (1,1)")

# 3. Horizontal line through (5, -2)
x3 = np.linspace(0, 10, 100)
y3 = np.full_like(x3, -2)
axs[2].plot(x3, y3, label='Slope = 0', color='orange')
axs[2].scatter([5], [-2], color='black')
axs[2].set_title("Horizontal line through (5, -2)")

# 4. Vertical line through (-1, 3)
y4 = np.linspace(0, 6, 100)
x4 = np.full_like(y4, -1)
axs[3].plot(x4, y4, label='Undefined slope', color='red')
axs[3].scatter([-1], [3], color='black')
axs[3].set_title("Vertical line through (-1, 3)")

# 5. Decreases 3 units vertically for every 2 units it goes right
x5 = np.array([0, 2])
y5 = np.array([0, -3])
axs[4].plot(x5, y5, label='Slope = -3/2', color='purple')
axs[4].scatter(x5, y5, color='black')
axs[4].set_title("Slope = -3/2")

# Format all plots
for ax in axs[:5]:
    ax.axhline(0, color='gray', linewidth=0.5)
    ax.axvline(0, color='gray', linewidth=0.5)
    ax.grid(True)
    ax.set_xlim(-5, 10)
    ax.set_ylim(-10, 10)
    ax.legend()

# Hide unused 6th subplot
axs[5].axis('off')

plt.tight_layout()
plt.show()
```
