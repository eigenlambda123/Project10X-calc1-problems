### **Problems**

Given that $\sin(a) = \dfrac{3}{5}$ and $\cos(b) = \dfrac{5}{13}$, both in Quadrant I:

**14.** Find $\sin(a + b)$  
**15.** Find $\cos(a - b)$  
**16.** Find $\tan(a + b)$

---
### **Solutions**

### Step 1: Build the triangles

For $\sin(a) = \dfrac{3}{5}$:
- Opposite = 3, Hypotenuse = 5  
- So adjacent = $\sqrt{5^2 - 3^2} = \sqrt{25 - 9} = \sqrt{16} = 4$

Therefore:  
- $\cos(a) = \dfrac{4}{5}$

For $\cos(b) = \dfrac{5}{13}$:
- Adjacent = 5, Hypotenuse = 13  
- So opposite = $\sqrt{13^2 - 5^2} = \sqrt{169 - 25} = \sqrt{144} = 12$

Therefore:  
- $\sin(b) = \dfrac{12}{13}$  
- $\tan(a) = \dfrac{3}{4}$  
- $\tan(b) = \dfrac{12}{5}$

**14.** Use:
$$
\sin(a + b) = \sin(a)\cos(b) + \cos(a)\sin(b)
$$

Substitute:
$$
= \dfrac{3}{5} \cdot \dfrac{5}{13} + \dfrac{4}{5} \cdot \dfrac{12}{13}
= \dfrac{15 + 48}{65} = \dfrac{63}{65}
$$

**Answer:**  
$\boxed{\dfrac{63}{65}}$


**15.** Use:
$$
\cos(a - b) = \cos(a)\cos(b) + \sin(a)\sin(b)
$$

Substitute:
$$
= \dfrac{4}{5} \cdot \dfrac{5}{13} + \dfrac{3}{5} \cdot \dfrac{12}{13}
= \dfrac{20 + 36}{65} = \dfrac{56}{65}
$$

**Answer:**  
$\boxed{\dfrac{56}{65}}$


**16.** Use:
$$
\tan(a + b) = \dfrac{\tan(a) + \tan(b)}{1 - \tan(a)\tan(b)}
$$

Substitute:
- $\tan(a) = \dfrac{3}{4}$
- $\tan(b) = \dfrac{12}{5}$

Then:
$$
\tan(a + b) = \dfrac{\dfrac{3}{4} + \dfrac{12}{5}}{1 - \dfrac{3}{4} \cdot \dfrac{12}{5}} 
= \dfrac{\dfrac{15 + 48}{20}}{1 - \dfrac{36}{20}} 
= \dfrac{\dfrac{63}{20}}{\dfrac{-16}{20}} = \dfrac{63}{-16} = -\dfrac{63}{16}
$$

**Answer:**  
$\boxed{-\dfrac{63}{16}}$
