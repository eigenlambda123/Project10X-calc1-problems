### **Problems**

**1.** Evaluate: $\sin\left(75^\circ\right)$  
**2.** Evaluate: $\cos\left(15^\circ\right)$  
**3.** Evaluate: $\tan\left(105^\circ\right)$  
**4.** Evaluate: $\sin\left(\dfrac{\pi}{12}\right)$  
**5.** Evaluate: $\cos\left(\dfrac{5\pi}{12}\right)$

_Hint: Break each angle into a sum/difference of common reference angles: $30^\circ$, $45^\circ$, $60^\circ$ or $\dfrac{\pi}{6}, \dfrac{\pi}{4}, \dfrac{\pi}{3}$._

---

### **Solutions**


**1.** Break into:  
$$
75^\circ = 45^\circ + 30^\circ
$$

Use the identity:
$$
\sin(a + b) = \sin(a)\cos(b) + \cos(a)\sin(b)
$$

Substitute:
$$
\sin(75^\circ) = \sin(45^\circ)\cos(30^\circ) + \cos(45^\circ)\sin(30^\circ)
$$

Use exact values:
- $\sin(45^\circ) = \cos(45^\circ) = \dfrac{\sqrt{2}}{2}$
- $\sin(30^\circ) = \dfrac{1}{2}$, $\cos(30^\circ) = \dfrac{\sqrt{3}}{2}$

So:
$$
\sin(75^\circ) = \dfrac{\sqrt{2}}{2} \cdot \dfrac{\sqrt{3}}{2} + \dfrac{\sqrt{2}}{2} \cdot \dfrac{1}{2}
= \dfrac{\sqrt{6} + \sqrt{2}}{4}
$$

**Answer:**  
$\boxed{\dfrac{\sqrt{6} + \sqrt{2}}{4}}$


**2.** Use:
$$
\cos(a - b) = \cos(a)\cos(b) + \sin(a)\sin(b)
$$

Then:
$$
\cos(15^\circ) = \cos(45^\circ)\cos(30^\circ) + \sin(45^\circ)\sin(30^\circ)
$$

Substitute:
$$
= \dfrac{\sqrt{2}}{2} \cdot \dfrac{\sqrt{3}}{2} + \dfrac{\sqrt{2}}{2} \cdot \dfrac{1}{2}
= \dfrac{\sqrt{6} + \sqrt{2}}{4}
$$

**Answer:**  
$\boxed{\dfrac{\sqrt{6} + \sqrt{2}}{4}}$


**3.** Use:
$$
\tan(a + b) = \dfrac{\tan(a) + \tan(b)}{1 - \tan(a)\tan(b)}
$$

Values:
- $\tan(60^\circ) = \sqrt{3}$
- $\tan(45^\circ) = 1$

So:
$$
\tan(105^\circ) = \dfrac{\sqrt{3} + 1}{1 - \sqrt{3} \cdot 1}
= \dfrac{\sqrt{3} + 1}{1 - \sqrt{3}}
$$

Rationalize:
Multiply numerator and denominator by $1 + \sqrt{3}$:
$$
\tan(105^\circ) = \dfrac{(\sqrt{3} + 1)^2}{(1 - \sqrt{3})(1 + \sqrt{3})}
= \dfrac{3 + 2\sqrt{3} + 1}{1 - 3}
= \dfrac{4 + 2\sqrt{3}}{-2} = -2 - \sqrt{3}
$$

**Answer:**  
$\boxed{-2 - \sqrt{3}}$


**4.** Use:
$$
\sin(a - b) = \sin(a)\cos(b) - \cos(a)\sin(b)
$$

Substitute:
- $\sin\left(\dfrac{\pi}{4}\right) = \dfrac{\sqrt{2}}{2}$
- $\cos\left(\dfrac{\pi}{6}\right) = \dfrac{\sqrt{3}}{2}$
- $\cos\left(\dfrac{\pi}{4}\right) = \dfrac{\sqrt{2}}{2}$
- $\sin\left(\dfrac{\pi}{6}\right) = \dfrac{1}{2}$

Then:
$$
\sin\left(\dfrac{\pi}{12}\right) = \dfrac{\sqrt{2}}{2} \cdot \dfrac{\sqrt{3}}{2} - \dfrac{\sqrt{2}}{2} \cdot \dfrac{1}{2}
= \dfrac{\sqrt{6} - \sqrt{2}}{4}
$$

**Answer:**  
$\boxed{\dfrac{\sqrt{6} - \sqrt{2}}{4}}$


**5.** Use:
$$
\cos(a + b) = \cos(a)\cos(b) - \sin(a)\sin(b)
$$

Substitute:
- $\cos\left(\dfrac{\pi}{4}\right) = \dfrac{\sqrt{2}}{2}$  
- $\cos\left(\dfrac{\pi}{6}\right) = \dfrac{\sqrt{3}}{2}$  
- $\sin\left(\dfrac{\pi}{4}\right) = \dfrac{\sqrt{2}}{2}$  
- $\sin\left(\dfrac{\pi}{6}\right) = \dfrac{1}{2}$

Then:
$$
\cos\left(\dfrac{5\pi}{12}\right) = \dfrac{\sqrt{2}}{2} \cdot \dfrac{\sqrt{3}}{2} - \dfrac{\sqrt{2}}{2} \cdot \dfrac{1}{2}
= \dfrac{\sqrt{6} - \sqrt{2}}{4}
$$

**Answer:**  
$\boxed{\dfrac{\sqrt{6} - \sqrt{2}}{4}}$
