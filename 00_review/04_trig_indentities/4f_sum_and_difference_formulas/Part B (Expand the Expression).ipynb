### **Problems**

**6.** Expand: $\sin(x + y)$  
**7.** Expand: $\cos(x - y)$  
**8.** Expand: $\tan(a + b)$  
**9.** Expand: $\cos(2x - y)$  
**10.** Expand: $\tan(x - \frac{\pi}{4})$

### **Solutions**


**6.** **Formula:**  
$$
\sin(a + b) = \sin(a)\cos(b) + \cos(a)\sin(b)
$$

**Answer:**  
$$
\boxed{\sin(x)\cos(y) + \cos(x)\sin(y)}
$$


**7.** **Formula:**  
$$
\cos(a - b) = \cos(a)\cos(b) + \sin(a)\sin(b)
$$

**Answer:**  
$$
\boxed{\cos(x)\cos(y) + \sin(x)\sin(y)}
$$


**8.** **Formula:**  
$$
\tan(a + b) = \dfrac{\tan(a) + \tan(b)}{1 - \tan(a)\tan(b)}
$$

**Answer:**  
$$
\boxed{\dfrac{\tan(a) + \tan(b)}{1 - \tan(a)\tan(b)}}
$$


**9.** Use the formula:  
$$
\cos(a - b) = \cos(a)\cos(b) + \sin(a)\sin(b)
$$  
with $a = 2x$, $b = y$

**Answer:**  
$$
\boxed{\cos(2x)\cos(y) + \sin(2x)\sin(y)}
$$


**10.** Formula:**  
$$
\tan(a - b) = \dfrac{\tan(a) - \tan(b)}{1 + \tan(a)\tan(b)}
$$

Use:
- $\tan\left(\dfrac{\pi}{4}\right) = 1$

Then:
$$
\tan\left(x - \dfrac{\pi}{4}\right) = \dfrac{\tan(x) - 1}{1 + \tan(x)}
$$

**Answer:**  
$$
\boxed{\dfrac{\tan(x) - 1}{1 + \tan(x)}}
$$
