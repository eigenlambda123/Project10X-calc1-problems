### **Problems**

Evaluate the following without a calculator:

1. $\sin^{-1}(1)$
2. $\cos^{-1}\left(\frac{1}{2}\right)$
3. $\tan^{-1}(1)$
4. $\cos^{-1}(-1)$
5. $\sin^{-1}\left(-\frac{\sqrt{2}}{2}\right)$
6. $\tan^{-1}(\sqrt{3})$
7. $\sin^{-1}(0)$
8. $\cos^{-1}(0)$

---
### **Solutions**

**1.** $\sin^{-1}(1)$

We are looking for the angle $\theta$ such that:

$$
\sin(\theta) = 1
$$

within the principal range of $\sin^{-1}(x)$:

$$
\left[-\frac{\pi}{2}, \frac{\pi}{2}\right]
$$

So,

$$
\sin^{-1}(1) = \boxed{\frac{\pi}{2}}
$$


**2.** $\cos^{-1}\left(\frac{1}{2}\right)$

We are looking for $\theta$ such that:

$$
\cos(\theta) = \frac{1}{2}
$$

within the range of $\cos^{-1}(x)$:

$$
[0, \pi]
$$

So,

$$
\cos^{-1}\left(\frac{1}{2}\right) = \boxed{\frac{\pi}{3}}
$$
