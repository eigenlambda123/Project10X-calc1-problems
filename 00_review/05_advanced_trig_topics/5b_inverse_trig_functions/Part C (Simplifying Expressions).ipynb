### **Problems**

Simplify the following expressions (assume $x$ is in the domain of the expression):

1. $\sin(\sin^{-1}(x))$
2. $\cos(\cos^{-1}(x))$
3. $\tan(\tan^{-1}(x))$
4. $\sin^{-1}(\sin(x))$
5. $\tan^{-1}(\tan(x))$
6. $\cos^{-1}(\cos(x))$

---
### **Solutions**

**1.** $\sin(\sin^{-1}(x))$

This is a direct composition. Since $\sin^{-1}(x)$ returns a value in $\left[-\frac{\pi}{2}, \frac{\pi}{2}\right]$ and $x \in [-1, 1]$, we simply get:

$$
\sin(\sin^{-1}(x)) = \boxed{x}
$$


**2.** $\cos(\cos^{-1}(x))$

Same logic: $\cos^{-1}(x)$ returns a value in $[0, \pi]$, and $x \in [-1, 1]$, so:

$$
\cos(\cos^{-1}(x)) = \boxed{x}
$$


**3.** $\tan(\tan^{-1}(x))$

$\tan^{-1}(x)$ has a range of $\left(-\frac{\pi}{2}, \frac{\pi}{2}\right)$, and for all real $x$:

$$
\tan(\tan^{-1}(x)) = \boxed{x}
$$


**4.** $\sin^{-1}(\sin(x))$

This is **not always** $x$ unless $x \in \left[-\frac{\pi}{2}, \frac{\pi}{2}\right]$, the principal range of $\sin^{-1}(x)$.

So:

$$
\sin^{-1}(\sin(x)) = 
\begin{cases}
\boxed{x}, & \text{if } x \in \left[-\frac{\pi}{2}, \frac{\pi}{2}\right] \\
\text{a value in } \left[-\frac{\pi}{2}, \frac{\pi}{2}\right] \text{ with same sine as } x, & \text{otherwise}
\end{cases}
$$


**5.** $\tan^{-1}(\tan(x))$

Same idea: this is only equal to $x$ if $x \in \left(-\frac{\pi}{2}, \frac{\pi}{2}\right)$.

$$
\tan^{-1}(\tan(x)) = 
\begin{cases}
\boxed{x}, & \text{if } x \in \left(-\frac{\pi}{2}, \frac{\pi}{2}\right) \\
\text{a value in } \left(-\frac{\pi}{2}, \frac{\pi}{2}\right) \text{ with same tangent as } x, & \text{otherwise}
\end{cases}
$$


**6.** $\cos^{-1}(\cos(x))$

$\cos^{-1}(x)$ has a range of $[0, \pi]$, so this expression is only equal to $x$ if $x \in [0, \pi]$.

$$
\cos^{-1}(\cos(x)) = 
\begin{cases}
\boxed{x}, & \text{if } x \in [0, \pi] \\
\boxed{2\pi - x} \text{ or another equivalent angle in } [0, \pi], & \text{otherwise}
\end{cases}
$$
