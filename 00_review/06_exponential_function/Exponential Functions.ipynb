# **Exponential Functions**

### **Key Concepts**

An **exponential function** has the form:  
$$f(x) = a^x$$
Where:

- $a > 0$, $a \ne 1$
- $x$ is a real number in the exponent

### **Key Properties**

- **Domain**: $(-\infty, \infty)$
- **Range**: $(0, \infty)$
- **Intercept**: $(0, 1)$
- **Horizontal Asymptote**: $y = 0$
- **Growth** if $a > 1$
- **Decay** if $0 < a < 1$

### **Laws of Exponents**

- $a^x \cdot a^y = a^{x+y}$
- $\dfrac{a^x}{a^y} = a^{x - y}$
- $(a^x)^y = a^{xy}$
- $a^{-x} = \dfrac{1}{a^x}$
- $a^0 = 1$


### **Inverse Relationship**

- The inverse of $f(x) = a^x$ is:  $$f^{-1}(x) = \log_a(x)$$

- So:
    - $a^{\log_a(x)} = x$
    - $\log_a(a^x) = x$

### **Graphs**

```python
import numpy as np
import matplotlib.pyplot as plt

# Define x/t/z ranges
x = np.linspace(-2, 2, 1000)
t = np.linspace(-2, 5, 1000)
z = np.linspace(-2, 2, 1000)

fig, axes = plt.subplots(4, 1, figsize=(8, 10), sharex=False)
fig.suptitle("Graphing Functions", fontsize=14)

# 1. Exponential Function: f(x) = 3^(1+2x)
axes[0].plot(x, 3**(1 + 2 * x), color='blue')
axes[0].set_title(r"$f(x) = 3^{1 + 2x}$")
axes[0].set_ylim(0, 100)
axes[0].axhline(0, color='black', linewidth=0.5)
axes[0].grid(True)

# 2. Linear Function: h(x) = 1 - x/4
axes[1].plot(x, 1 - x/4, color='green')
axes[1].set_title(r"$h(x) = 1 - \frac{x}{4}$")
axes[1].axhline(0, color='black', linewidth=0.5)
axes[1].grid(True)

# 3. Exponential Growth: h(t) = 8 + 3e^{2t - 4}
axes[2].plot(t, 8 + 3 * np.exp(2 * t - 4), color='red')
axes[2].set_title(r"$h(t) = 8 + 3e^{2t - 4}$")
axes[2].axhline(y=8, color='gray', linestyle='--', linewidth=1, label='Asymptote: y=8')
axes[2].legend()
axes[2].grid(True)

# 4. Exponential Decay: g(z) = 10 - (1/4)e^{-2 - 3z}
axes[3].plot(z, 10 - (1/4) * np.exp(-2 - 3 * z), color='purple')
axes[3].set_title(r"$g(z) = 10 - \frac{1}{4}e^{-2 - 3z}$")
axes[3].axhline(y=10, color='gray', linestyle='--', linewidth=1, label='Asymptote: y=10')
axes[3].legend()
axes[3].grid(True)

# Final formatting
for ax in axes:
    ax.label_outer()

plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()
```
