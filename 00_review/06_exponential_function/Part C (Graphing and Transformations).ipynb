### **Problems**

**Describe the transformation and sketch:**
1. $f(x) = 2^x$  
2. $f(x) = 2^{x-1} + 3$  
3. $f(x) = -3^x$  
4. $f(x) = \left(\frac{1}{2}\right)^x$  
5. What is the horizontal asymptote of $f(x) = 4^x - 5$?

---

### **Solutions**

**1.** $f(x) = 2^x$

- **Base Function:** Exponential growth.
- **Transformations:** None.
- **Graph Shape:** Increases rapidly as $x \to \infty$; approaches 0 as $x \to -\infty$.
- **Horizontal Asymptote:** $$\boxed{y = 0}$$


**2.** $f(x) = 2^{x - 1} + 3$

- **Base Function:** $2^x$
- **Transformations:**
  - Shift **right** by 1 unit (due to $x - 1$).
  - Shift **up** by 3 units.
- **Horizontal Asymptote:** Shifted from $y = 0$ to $$\boxed{y = 3}$$


**3.** $f(x) = -3^x$

- **Base Function:** $3^x$ (exponential growth).
- **Transformations:**
  - Reflect over the **x-axis** (due to the negative sign).
- **Graph Shape:** Decreases rapidly as $x \to \infty$; approaches 0 from below as $x \to -\infty$.
- **Horizontal Asymptote:** $$\boxed{y = 0}$$


**4.** $f(x) = \left(\frac{1}{2}\right)^x$

- **Base Function:** Exponential **decay**.
- **Graph Shape:** Decreases as $x \to \infty$, increases toward 1 as $x \to -\infty$.
- **Horizontal Asymptote:** $$\boxed{y = 0}$$


**5.** What is the horizontal asymptote of $f(x) = 4^x - 5$?

- **Base Function:** $4^x$, with a **downward shift** of 5 units.
- **Horizontal Asymptote:** $$\boxed{y = -5}$$
- **Transformations:** None.
- **Graph Shape:** Increases rapidly as $x \to \infty$; approaches 0 as $x \to -\infty$.
- **Horizontal Asymptote:** $$\boxed{y = 0}$$


---
### **Visualizations**

```python
import numpy as np
import matplotlib.pyplot as plt

# Define x values
x = np.linspace(-5, 5, 400)

# Define functions
y1 = 2**x
y2 = 2**(x - 1) + 3
y3 = -3**x
y4 = (1/2)**x
y5 = 4**x - 5

# Set up subplots
fig, axs = plt.subplots(3, 2, figsize=(12, 12))
axs = axs.flatten()

# Plot each function with title and horizontal asymptote
axs[0].plot(x, y1, label=r"$f(x) = 2^x$")
axs[0].axhline(0, color='gray', linestyle='--', label='Asymptote: $y=0$')
axs[0].set_title(r"$f(x) = 2^x$")

axs[1].plot(x, y2, label=r"$f(x) = 2^{x-1} + 3$")
axs[1].axhline(3, color='gray', linestyle='--', label='Asymptote: $y=3$')
axs[1].set_title(r"$f(x) = 2^{x-1} + 3$")

axs[2].plot(x, y3, label=r"$f(x) = -3^x$")
axs[2].axhline(0, color='gray', linestyle='--', label='Asymptote: $y=0$')
axs[2].set_title(r"$f(x) = -3^x$")

axs[3].plot(x, y4, label=r"$f(x) = \left(\frac{1}{2}\right)^x$")
axs[3].axhline(0, color='gray', linestyle='--', label='Asymptote: $y=0$')
axs[3].set_title(r"$f(x) = \left(\frac{1}{2}\right)^x$")

axs[4].plot(x, y5, label=r"$f(x) = 4^x - 5$")
axs[4].axhline(-5, color='gray', linestyle='--', label='Asymptote: $y=-5$')
axs[4].set_title(r"$f(x) = 4^x - 5$")

# Hide the last subplot
axs[5].axis('off')

# General formatting
for ax in axs[:5]:
    ax.legend()
    ax.grid(True)
    ax.set_ylim(-10, 20)
    ax.set_xlim(-5, 5)
    ax.axvline(0, color='black', linewidth=0.5)
    ax.axhline(0, color='black', linewidth=0.5)

fig.suptitle("Exponential Function Transformations", fontsize=16)
plt.tight_layout()
plt.show()
```
