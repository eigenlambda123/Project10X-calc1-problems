### **Problems**
**Problem 1**

Simplify:

$$
\log_2(8 \cdot 4)
$$


**Problem 2**

Simplify:

$$
\log_{10}(5x)
$$


**Problem 3**

Simplify:

$$
\ln(3xy)
$$


**Problem 4**

Simplify:

$$
\log_b(7 \cdot a \cdot x)
$$


**Problem 5**

True or False?

$$
\log_b(2x^2) = \log_b 2 + \log_b x^2
$$


**Problem 6**

Rewrite the expression using log addition:

$$
\log_3(9m^2n)
$$


**Problem 7**

Simplify and evaluate:

$$
\log_{10}(10 \cdot 100)
$$


**Problem 8**

Let $\log_5 x = 2$ and $\log_5 y = 3$. Find:

$$
\log_5(xy)
$$


**Problem 9**

Simplify fully:

$$
\ln(2x^3y)
$$


**Problem 10**

Given:

- $\log_7 a = 2$
- $\log_7 b = 4$
- $\log_7 c = 1$

Find:

$$
\log_7(abc)
$$

---

### **Solutions**

**1.** $\log_2(8 \cdot 4) = \log_2 8 + \log_2 4 = 3 + 2 = \boxed{5}$

**2.** $\log_{10}(5x) = \log_{10} 5 + \log_{10} x = \boxed{\log 5 + \log x}$

**3.** $\ln(3xy) = \ln 3 + \ln x + \ln y = \boxed{\ln 3 + \ln x + \ln y}$

**4.** $\log_b(7 \cdot a \cdot x) = \log_b 7 + \log_b a + \log_b x = \boxed{\log_b 7 + \log_b a + \log_b x}$

**5.** $\log_b(2x^2) = \log_b 2 + \log_b x^2$ â€”  **True**

**6.** $\log_3(9m^2n) = \log_3 9 + \log_3 m^2 + \log_3 n = 2 + 2\log_3 m + \log_3 n = \boxed{2 + 2\log_3 m + \log_3 n}$

**7.** $\log_{10}(10 \cdot 100) = \log_{10} 10 + \log_{10} 100 = 1 + 2 = \boxed{3}$

**8.** $\log_5(xy) = \log_5 x + \log_5 y = 2 + 3 = \boxed{5}$

**9.** $\ln(2x^3y) = \ln 2 + \ln x^3 + \ln y = \ln 2 + 3\ln x + \ln y = \boxed{\ln 2 + 3\ln x + \ln y}$

**10.** $\log_7(abc) = \log_7 a + \log_7 b + \log_7 c = 2 + 4 + 1 = \boxed{7}$
