### **Problems**

**Problem 1**

Simplify:

$$
\log_{10}\left(\frac{100}{10}\right)
$$


**Problem 2**

Simplify:

$$
\ln\left(\frac{x}{3}\right)
$$


**Problem 3**

Simplify:

$$
\log_b\left(\frac{a^3}{b^2}\right)
$$


**Problem 4**

Rewrite using subtraction of logs:

$$
\log_4\left(\frac{16m^2}{n}\right)
$$


**Problem 5**

True or False?

$$
\log_b\left(\frac{x^2y}{z^3}\right) = \log_b x^2 + \log_b y - \log_b z^3
$$


**Problem 6**

Evaluate:

$$
\log_2\left(\frac{32}{4}\right)
$$


**Problem 7**

Let $\log_3 x = 5$ and $\log_3 y = 2$. Find:

$$
\log_3\left(\frac{x}{y}\right)
$$


**Problem 8**

Simplify fully:

$$
\ln\left(\frac{5x^3}{y^2}\right)
$$


**Problem 9**

Given:

- $\log_6 a = 1$
- $\log_6 b = 2$

Find:

$$
\log_6\left(\frac{a^2}{b^3}\right)
$$


**Problem 10**

Simplify:

$$
\log_b\left(\frac{7a^4x^2}{9y}\right)
$$

---


### **Solutions**

**1.** $\log_{10}\left(\frac{100}{10}\right) = \log_{10} 100 - \log_{10} 10 = 2 - 1 = \boxed{1}$

**2.** $\ln\left(\frac{x}{3}\right) = \ln x - \ln 3 = \boxed{\ln x - \ln 3}$

**3.** $\log_b\left(\frac{a^3}{b^2}\right) = \log_b a^3 - \log_b b^2 = 3\log_b a - 2\log_b b = \boxed{3\log_b a - 2\log_b b}$

**4.** $\log_4\left(\frac{16m^2}{n}\right) = \log_4 16 + \log_4 m^2 - \log_4 n = 2 + 2\log_4 m - \log_4 n = \boxed{2 + 2\log_4 m - \log_4 n}$

**5.** $\log_b\left(\frac{x^2y}{z^3}\right) = \log_b x^2 + \log_b y - \log_b z^3$ â€” **True**

**6.** $\log_2\left(\frac{32}{4}\right) = \log_2 32 - \log_2 4 = 5 - 2 = \boxed{3}$

**7.** $\log_3\left(\frac{x}{y}\right) = \log_3 x - \log_3 y = 5 - 2 = \boxed{3}$

**8.** $\ln\left(\frac{5x^3}{y^2}\right) = \ln 5 + \ln x^3 - \ln y^2 = \ln 5 + 3\ln x - 2\ln y = \boxed{\ln 5 + 3\ln x - 2\ln y}$

**9.** $\log_6\left(\frac{a^2}{b^3}\right) = 2\log_6 a - 3\log_6 b = 2(1) - 3(2) = 2 - 6 = \boxed{-4}$

**10.** $\log_b\left(\frac{7a^4x^2}{9y}\right) = \log_b 7 + \log_b a^4 + \log_b x^2 - \log_b 9 - \log_b y = \boxed{\log_b 7 + 4\log_b a + 2\log_b x - \log_b 9 - \log_b y}$
