### **Problems**

**Problem 1**
Simplify using the Change of Base Rule:
$$
\log_2 16
$$


**Problem 2**
Evaluate using common logarithms:
$$
\log_5 125
$$


**Problem 3**
Evaluate using natural logarithms:
$$
\log_3 81
$$


**Problem 4**
Evaluate and round to 2 decimal places:
$$
\log_7 100
$$


**Problem 5**
Use the Change of Base Rule to express:
$$
\log_{10} 1000
$$

using base $e$


**Problem 6**
Rewrite the expression using Change of Base Rule:
$$
\log_b a^2
$$


**Problem 7**
Rewrite the expression using natural logs:
$$
\log_2 x
$$


**Problem 8**
Simplify:
$$
\log_4 20
$$

using base 10 logs


**Problem 9**
Evaluate and round to 2 decimal places:
$$
\log_6 50
$$


**Problem 10**
True or False?

$$
\log_3 9 = \frac{\ln 9}{\ln 3}
$$

---


### **Solutions**

**1.**  $\log_2 16 = \dfrac{\log 16}{\log 2} = \dfrac{4}{1} = \boxed{4}$


**2.**  $\log_5 125 = \dfrac{\log 125}{\log 5} = \dfrac{3}{1} = \boxed{3}$  
(because $125 = 5^3$)


**3.**  $\log_3 81 = \dfrac{\ln 81}{\ln 3} = \dfrac{4 \ln 3}{\ln 3} = \boxed{4}$  
(because $81 = 3^4$)


**4.**  $\log_7 100 = \dfrac{\log 100}{\log 7} \approx \dfrac{2}{0.8451} \approx \boxed{2.37}$


**5.**  $\log_{10} 1000 = \dfrac{\ln 1000}{\ln 10} = \dfrac{3 \ln 10}{\ln 10} = \boxed{3}$


**6.**  $\log_b a^2 = 2 \log_b a = 2 \cdot \dfrac{\log a}{\log b} = \boxed{\dfrac{2 \log a}{\log b}}$


**7.**  $\log_2 x = \dfrac{\ln x}{\ln 2} = \boxed{\dfrac{\ln x}{\ln 2}}$


**8.**  $\log_4 20 = \dfrac{\log 20}{\log 4} \approx \dfrac{1.3010}{0.6021} \approx \boxed{2.16}$


**9.**  $\log_6 50 = \dfrac{\log 50}{\log 6} \approx \dfrac{1.6989}{0.7782} \approx \boxed{2.18}$


**10.**  $\log_3 9 = 2$ and  $\dfrac{\ln 9}{\ln 3} = \dfrac{2 \ln 3}{\ln 3} = 2$  So the statement is **$\boxed{True}$**
