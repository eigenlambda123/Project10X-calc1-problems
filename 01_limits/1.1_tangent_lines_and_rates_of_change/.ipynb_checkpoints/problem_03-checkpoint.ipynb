### **Problems**

Given the function:
$$ W(x) = \ln(1 + x^4) $$

And the point:
$$ P \text{ at } x = 1 $$

### **Part (a): Secant Line Slopes**

For points Q with the following x-values (approximate slopes to 8 decimal places), compute the slope ($m_{PQ}$) of the secant line through P and Q:

#### Right-side Points (approaching from above):

(i) $x = 1.5$  
(ii) $x = 1.1$  
(iii) $x = 1.01$  
(iv) $x = 1.001$  
(v) $x = 1.0001$

#### Left-side Points (approaching from below):

(vi) $x = 0.5$  
(vii) $x = 0.9$  
(viii) $x = 0.99$  
(ix) $x = 0.999$  
(x) $x = 0.9999$

### **Part (b): Tangent Line Estimation**

Using results from part (a):
1. Estimate the slope of the tangent line at $x = 1$
2. Write the equation of the tangent line

---


### **Solutions**

### **Part (a): Secant Line Slopes**

**Slope Formula**:
$$ m_{PQ} = \frac{W(x_Q) - W(1)}{x_Q - 1} = \frac{\ln(1+x_Q^4) - \ln2}{x_Q - 1} $$

#### Right-side Approximations (x > 1):

| x_Q  | W(x_Q)               | Slope (m_{PQ})       |
|------|----------------------|----------------------|
| 1.5  | $\ln(6.0625)≈1.80171$ | $\frac{1.80171-0.69315}{0.5}≈2.21712$ |
| 1.1  | $\ln(2.4641)≈0.90198$ | $\frac{0.90198-0.69315}{0.1}≈2.08830$ |
| 1.01 | $\ln(2.04060401)≈0.71335$ | $\frac{0.71335-0.69315}{0.01}≈2.02000$ |
| 1.001| $\ln(2.004006004001)≈0.69533$ | $\frac{0.69533-0.69315}{0.001}≈2.00200$ |
| 1.0001| $\ln(2.000400060004)≈0.69335$ | $\frac{0.69335-0.69315}{0.0001}≈2.00020$ |

#### Left-side Approximations (x < 1):

| x_Q  | W(x_Q)               | Slope (m_{PQ})       |
|------|----------------------|----------------------|
| 0.5  | $\ln(1.0625)≈0.06062$ | $\frac{0.06062-0.69315}{-0.5}≈1.26506$ |
| 0.9  | $\ln(1.6561)≈0.50424$ | $\frac{0.50424-0.69315}{-0.1}≈1.88910$ |
| 0.99 | $\ln(1.96059601)≈0.67333$ | $\frac{0.67333-0.69315}{-0.01}≈1.98200$ |
| 0.999| $\ln(1.996005996001)≈0.69081$ | $\frac{0.69081-0.69315}{-0.001}≈1.99820$ |
| 0.9999| $\ln(1.999600059996)≈0.69253$ | $\frac{0.69253-0.69315}{-0.0001}≈1.99982$ |

### **Part (b): Tangent Line Estimation**

**Slope Estimation**:
As $x_Q \to 1$, slopes approach $\boxed{2}$ from both sides.

**Tangent Line Equation**:
Using point-slope form at $P(1, \ln2)$ with $m = 2$:
$$ y - \ln2 = 2(x - 1) $$
$$ \boxed{y = 2x - 2 + \ln2} $$

### **Visualization**

```python
import numpy as np
import matplotlib.pyplot as plt
from math import log

# Define the function
def W(x):
    return np.log(1 + x**4)

# Create plot
plt.figure(figsize=(10, 6))
x = np.linspace(0, 2, 400)  # x-range to show behavior around x=1
plt.plot(x, W(x), 'b-', label=r'$W(x) = \ln(1 + x^4)$', linewidth=2)

# Point P
x_p, y_p = 1, W(1)
plt.plot(x_p, y_p, 'ro', label=f'Point P (1, {y_p:.4f})')

# Example secant lines (left and right)
x_q_right = 1.5
x_q_left = 0.5
for x_q, color, side in [(x_q_right, 'g', 'right'), (x_q_left, 'm', 'left')]:
    m = (W(x_q) - W(x_p))/(x_q - x_p)
    secant = m*(x - x_p) + y_p
    plt.plot(x, secant, f'{color}--', 
             label=f'Secant ({side}) m={m:.2f}', alpha=0.7)

# Tangent line
m_tangent = 2
tangent = m_tangent*(x - x_p) + y_p
plt.plot(x, tangent, 'k-', label=f'Tangent m={m_tangent}', linewidth=2)

# Formatting
plt.title('Function with Secant and Tangent Lines at x=1')
plt.xlabel('x')
plt.ylabel('W(x)')
plt.grid(True, alpha=0.3)
plt.legend()
plt.xlim(0, 2)  # Adjusted x-range
plt.ylim(0, 2.5) # Adjusted y-range
plt.xticks(np.arange(0, 2.1, 0.25))
plt.tight_layout()
plt.show()
```
