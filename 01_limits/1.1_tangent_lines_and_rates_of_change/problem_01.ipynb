### **Problem**

Given the function:
$$ f(x) = 3(x + 2)^2 $$

And the point:
$$ P \text{ at } x = -3 $$

### **Part (a): Secant Line Slopes**

For points Q with the following x-values (approximate to 8 decimal places), compute the slope (mₚQ) of the secant line through P and Q:

#### Left-side Points:

(i) $x = -3.5$  
(ii) $x = -3.1$  
(iii) $x = -3.01$  
(iv) $x = -3.001$  
(v) $x = -3.0001$

#### Right-side Points:

(vi) $x = -2.5$  
(vii) $x = -2.9$  
(viii) $x = -2.99$  
(ix) $x = -2.999$  
(x) $x = -2.9999$

### **Part (b): Tangent Line Estimation**

Using results from part (a):
1. Estimate the slope of the tangent line at $x = -3$
2. Write the equation of the tangent line

---

### **Solutions**

### **Part (a): Secant Line Slopes**

**Given**:  
- $f(x) = 3(x + 2)^2$
- Point P at $x = -3$ → $P(-3, f(-3)) = (-3, 3)$

**Slope formula**:  
$m_{PQ} = \frac{f(x_Q) - f(-3)}{x_Q - (-3)} = \frac{3(x_Q+2)^2 - 3}{x_Q + 3}$


#### Left-side Approximations:
| x_Q  | f(x_Q) | Slope (m_{PQ}) |
|------|--------|----------------|
| -3.5 | 0.75   | -4.5 |
| -3.1 | 2.43   | -5.7 |
| -3.01 | 2.9403 | -5.97 |
| -3.001 | 2.994003 | -5.997 |
| -3.0001 | 2.99940003 | -5.9997 |

#### Right-side Approximations:
| x_Q  | f(x_Q) | Slope (m_{PQ}) |
|------|--------|----------------|
| -2.5 | 0.75   | -4.5 |
| -2.9 | 3.63   | -6.3 |
| -2.99 | 3.0603 | -6.03 |
| -2.999 | 3.006003 | -6.003 |
| -2.9999 | 3.00060003 | -6.0003 |



### **Part (b): Tangent Line Estimation**

**Slope Estimation**:  
As $x_Q \to -3$, slopes approach $\boxed{6}$ from both sides.

**Tangent Line Equation**:  
Using point-slope form at $P(-3, 3)$:
$$ y - 3 = -6(x - (-3)) $$
$$ \boxed{y = -6x - 15} $$

---

### **Visualization**

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the function
def f(x):
    return 3*(x + 2)**2

# Create plot
plt.figure(figsize=(10, 6))
x = np.linspace(-4, -2, 400)
plt.plot(x, f(x), 'b-', label=r'$f(x) = 3(x+2)^2$', linewidth=2)

# Point P
x_p, y_p = -3, f(-3)
plt.plot(x_p, y_p, 'ro', label=f'Point P (-3, {y_p})')

# Example secant lines (left and right)
x_q_left = -3.5
x_q_right = -2.5
for x_q, color, side in [(x_q_left, 'g', 'left'), (x_q_right, 'm', 'right')]:
    m = (f(x_q) - f(x_p))/(x_q - x_p)
    secant = m*(x - x_p) + y_p
    plt.plot(x, secant, f'{color}--', 
             label=f'Secant ({side}) m={m:.1f}', alpha=0.7)

# Tangent line 
m_tangent = -6
tangent = m_tangent*(x - x_p) + y_p
plt.plot(x, tangent, 'k-', label=f'Tangent m={m_tangent}', linewidth=2)

# Formatting
plt.title('Function with Secant and Tangent Lines at x=-3')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True, alpha=0.3)
plt.legend()
plt.xlim(-4, -2)
plt.ylim(0, 10)
plt.xticks(np.arange(-4, -1, 0.5))
plt.tight_layout()
plt.show()
```
