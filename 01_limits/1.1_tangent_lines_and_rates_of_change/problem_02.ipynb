### **Problem**

Given the function:
$$ g(x) = \sqrt{4x + 8} $$

And the point:
$$ P \text{ at } x = 2 $$

### **Part (a): Secant Line Slopes**

For points Q with the following x-values (approximate slopes to 8 decimal places), compute the slope ($m_{PQ}$) of the secant line through P and Q:

#### Right-side Points (approaching from above):

(i) $x = 2.5$  
(ii) $x = 2.1$  
(iii) $x = 2.01$  
(iv) $x = 2.001$  
(v) $x = 2.0001$

#### Left-side Points (approaching from below):

(vi) $x = 1.5$  
(vii) $x = 1.9$  
(viii) $x = 1.99$  
(ix) $x = 1.999$  
(x) $x = 1.9999$

### **Part (b): Tangent Line Estimation**

Using results from part (a):
1. Estimate the slope of the tangent line at $x = 2$
2. Write the equation of the tangent line


---

### **Solutions**

### **Part (a): Secant Line Slopes**

**Slope Formula**:
$$ m_{PQ} = \frac{g(x_Q) - g(2)}{x_Q - 2} = \frac{\sqrt{4x_Q + 8} - 4}{x_Q - 2} $$

#### Right-side Approximations (x > 2):

| x_Q  | g(x_Q)     | Slope (m_{PQ})       |
|------|------------|----------------------|
| 2.5  | $\sqrt{18} ≈ 4.24264069$ | $\frac{4.24264069-4}{0.5} ≈ 0.48528137$ |
| 2.1  | $\sqrt{16.4} ≈ 4.04969135$ | $\frac{4.04969135-4}{0.1} ≈ 0.49691346$ |
| 2.01 | $\sqrt{16.04} ≈ 4.00499376$ | $\frac{4.00499376-4}{0.01} ≈ 0.49968752$ |
| 2.001| $\sqrt{16.004} ≈ 4.00049994$ | $\frac{4.00049994-4}{0.001} ≈ 0.49996875$ |
| 2.0001| $\sqrt{16.0004} ≈ 4.00005000$ | $\frac{4.00005000-4}{0.0001} ≈ 0.49999688$ |

#### Left-side Approximations (x < 2):

| x_Q  | g(x_Q)     | Slope (m_{PQ})       |
|------|------------|----------------------|
| 1.5  | $\sqrt{14} ≈ 3.74165739$ | $\frac{3.74165739-4}{-0.5} ≈ 0.51668523$ |
| 1.9  | $\sqrt{15.6} ≈ 3.94968353$ | $\frac{3.94968353-4}{-0.1} ≈ 0.50316475$ |
| 1.99 | $\sqrt{15.96} ≈ 3.99499687$ | $\frac{3.99499687-4}{-0.01} ≈ 0.50031258$ |
| 1.999| $\sqrt{15.996} ≈ 3.99949994$ | $\frac{3.99949994-4}{-0.001} ≈ 0.50003125$ |
| 1.9999| $\sqrt{15.9996} ≈ 3.99995000$ | $\frac{3.99995000-4}{-0.0001} ≈ 0.50000312$ |

### **Part (b): Tangent Line Estimation**

**Slope Estimation**:
As $x_Q \to 2$, slopes approach $\boxed{0.5}$ from both sides.

**Tangent Line Equation**:
Using point-slope form at $P(2, 4)$ with $m = 0.5$:
$$ y - 4 = 0.5(x - 2) $$
$$ \boxed{y = 0.5x + 3} $$

### **Visualization**

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the function
def g(x):
    return np.sqrt(4*x + 8)

# Create plot
plt.figure(figsize=(10, 6))
x = np.linspace(-2, 4, 400)  # x-range to show full domain and behavior
plt.plot(x, g(x), 'b-', label=r'$g(x) = \sqrt{4x + 8}$', linewidth=2)

# Point P
x_p, y_p = 2, g(2)
plt.plot(x_p, y_p, 'ro', label=f'Point P (2, {y_p})')

# Example secant lines (left and right)
x_q_right = 2.5
x_q_left = 1.5
for x_q, color, side in [(x_q_right, 'g', 'right'), (x_q_left, 'm', 'left')]:
    m = (g(x_q) - g(x_p))/(x_q - x_p)
    secant = m*(x - x_p) + y_p
    plt.plot(x, secant, f'{color}--', 
             label=f'Secant ({side}) m={m:.2f}', alpha=0.7)

# Tangent line
m_tangent = 0.5
tangent = m_tangent*(x - x_p) + y_p
plt.plot(x, tangent, 'k-', label=f'Tangent m={m_tangent}', linewidth=2)

# Formatting
plt.title('Function with Secant and Tangent Lines at x=2')
plt.xlabel('x')
plt.ylabel('g(x)')
plt.grid(True, alpha=0.3)
plt.legend()
plt.xlim(-2, 4)  # Shows full domain (4x+8≥0 → x≥-2)
plt.ylim(0, 6)   # Appropriate y-range
plt.xticks(np.arange(-2, 5, 1))
plt.tight_layout()
plt.show()
```
