### **Problems**

Given the position function:
$$ s(t) = (8 - t)(t + 6)^{\frac{3}{2}} $$

### **Part (a): Average Velocity**

Compute (accurate to at least 8 decimal places) the average velocity between:
- Fixed point: $t_1 = 10$  
- Variable point: $t_2$ = following values:

#### Right-side Points ($t_2 > 10$):

(i) $t_2 = 10.5$  
(ii) $t_2 = 10.1$  
(iii) $t_2 = 10.01$  
(iv) $t_2 = 10.001$  
(v) $t_2 = 10.0001$

#### Left-side Points ($t_2 < 10$):

(vi) $t_2 = 9.5$  
(vii) $t_2 = 9.9$  
(viii) $t_2 = 9.99$  
(ix) $t_2 = 9.999$  
(x) $t_2 = 9.9999$

**Formula**:  
Average Velocity $= \frac{s(t_2) - s(10)}{t_2 - 10}$

**Note**: 
- Negative positions indicate left of zero position
- $s(10) = (8-10)(10+6)^{3/2} = (-2)(64) = -128$

### **Part (b): Instantaneous Velocity**

Using results from part (a):
1. Estimate the instantaneous velocity at $t = 10$
2. Determine if the object is:
   - Moving right (positive velocity)
   - Moving left (negative velocity) 
   - Stationary (zero velocity)

---

### **Solutions**

**Formula**:
$$\text{Average Velocity} = \frac{s(t_2) - s(10)}{t_2 - 10} = \frac{(8-t_2)(t_2+6)^{3/2} + 128}{t_2 - 10}$$

### **Part (a): Average Velocity Table**

| t > 10 | A.V.          | t < 10 | A.V.          |
|--------|---------------|--------|---------------|
| 10.5   | -79.11658419  | 9.5    | -72.92931693  |
| 10.1   | -76.61966704  | 9.9    | -75.38216890  |
| 10.01  | -76.06188418  | 9.99   | -75.93813418  |
| 10.001 | -76.00618759  | 9.999  | -75.99381259  |
| 10.0001| -76.00061875  | 9.9999 | -75.99938125  |

### **Part (b): Instantaneous Velocity**

1. **Numerical Estimate**:
   - Right limit: → -76.0
   - Left limit: → -76.0
   - ∴ Instantaneous velocity = $\boxed{-76.0}$

2. **Analytical Verification**:
   $$s'(t) = -(t+6)^{3/2} + \tfrac{3}{2}(8-t)(t+6)^{1/2}$$
   $$s'(10) = -64 + \tfrac{3}{2}(-2)(4) = \boxed{-76}$$

3. **Conclusion**:
   The object is moving LEFT (velocity = $\boxed{-76.0}$ units/time) at t=10.

### **Visualization**

```python
import numpy as np
import matplotlib.pyplot as plt

def s(t):
    return (8 - t) * (t + 6)**1.5  # Using 1.5 instead of 3/2 for computational precision

t_left = np.linspace(9, 10, 400)
t_right = np.linspace(10, 11, 400)
t0 = 10
s0 = s(t0)

plt.figure(figsize=(10, 5))
plt.plot(t_left, s(t_left), 'b-', label='Left approach')
plt.plot(t_right, s(t_right), 'g-', label='Right approach')
plt.plot(t0, s0, 'ro', label=f't={t0}')

# Add tangent line
tangent = s0 - 76*(np.linspace(9,11,100)-t0)
plt.plot(np.linspace(9,11,100), tangent, 'k--', label='Tangent (v=-76)')

plt.title(f'Position Function\n(Converges to v={-76:.1f} at t=10)')
plt.xlabel('Time (t)')
plt.ylabel('Position s(t)')
plt.legend()
plt.grid(True)
plt.show()
```
