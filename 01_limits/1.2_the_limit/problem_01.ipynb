### **Problems**

For the function $f(x) = \frac{o - x}{x^2 - 4}$ answer each of the following questions.
### **(a)** 

Evaluate the function at the following values of $x$ compute (accurate to at least 8 decimal places).

- (i) $2.5$  
- (ii) $2.1$  
- (iii) $2.01$  
- (iv) $2.001$  
- (v) $2.0001$  

- (vi) $1.5$  
- (vii) $1.9$

### **(b)**

Use the information from (a) to estimate the value of $\lim_{x \to 2} \frac{8 - x^3}{x^2 - 4}$.

---

### **Solutions**

### **Part (a): Evaluate the Function**

Given the function:
$$ f(x) = \frac{8 - x^3}{x^2 - 4} $$

#### Approach:

1. **Simplify the function** (optional but helpful):
   - Denominator: $x^2 - 4 = (x - 2)(x + 2)$
   - Numerator: $8 - x^3 = -(x^3 - 8) = -(x - 2)(x^2 + 2x + 4)$
   - Simplified form (for $x \neq 2$):
 $$ f(x) = -\frac{x^2 + 2x + 4}{x + 2} $$

2. **Compute $f(x)$ for each $x$:**

| $x$       | Calculation                                                                 | $f(x)$ (approximate) |
|-----------|-----------------------------------------------------------------------------|----------------------|
| **(i) 2.5**  | $-\frac{2.5^2 + 2(2.5) + 4}{2.5 + 2} = -\frac{6.25 + 5 + 4}{4.5}$          | -3.38888889          |
| **(ii) 2.1** | $-\frac{2.1^2 + 2(2.1) + 4}{4.1}$                                          | -3.07560976          |
| **(iii) 2.01** | $-\frac{2.01^2 + 2(2.01) + 4}{4.01}$                                      | -3.00750623          |
| **(iv) 2.001** | $-\frac{2.001^2 + 2(2.001) + 4}{4.001}$                                   | -3.00075006          |
| **(v) 2.0001** | $-\frac{2.0001^2 + 2(2.0001) + 4}{4.0001}$                                | -3.00007500          |
| **(vi) 1.5** | $-\frac{1.5^2 + 2(1.5) + 4}{3.5}$                                         | -2.92857143          |
| **(vii) 1.9** | $-\frac{1.9^2 + 2(1.9) + 4}{3.9}$                                         | -2.92564103          |
| **(viii) 1.99** | $-\frac{1.99^2 + 2(1.99) + 4}{3.99}$                                      | -2.99250627          |
| **(ix) 1.999** | $-\frac{1.999^2 + 2(1.999) + 4}{3.999}$                                   | -2.99925006          |
| **(x) 1.9999** | $-\frac{1.9999^2 + 2(1.9999) + 4}{3.9999}$                                | -2.99992500          |

#### Observations:

- As $x \to 2^+$ (right side), $f(x) \to -3$.
- As $x \to 2^-$ (left side), $f(x) \to -3$.


### **Part (b): Estimate the Limit**

Using the results from (a), we observe:
$$ \lim_{x \to 2} \frac{8 - x^3}{x^2 - 4} = -3 $$

#### Verification:

Using the simplified form:
$$ \lim_{x \to 2} -\frac{x^2 + 2x + 4}{x + 2} = -\frac{4 + 4 + 4}{4} = -3 $$

#### Conclusion:

The limit is $\boxed{-3}$.
