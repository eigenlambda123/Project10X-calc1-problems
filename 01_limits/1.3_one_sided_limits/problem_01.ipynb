### **Problems**


Below is the graph of $f(x)$. For each of the given points determine:

1. The value of $f(a)$
2. $\lim_{x \to a^-} f(x)$ (left-hand limit)
3. $\lim_{x \to a^+} f(x)$ (right-hand limit) 
4. $\lim_{x \to a} f(x)$ (two-sided limit)

If any of these quantities do not exist, clearly explain why.

#### Points to Evaluate:

- (a) $a = -4$
- (b) $a = -1$
- (c) $a = 2$
- (d) $a = 4$

### **Solutions**

### **(a) At $a = -4$**

- **Function value**:  
  $f(-4) = 3$  
  *(Defined point on graph)*
- **Left-hand limit**:  
  $\lim_{x \to -4^-} f(x) = 3$
- **Right-hand limit**:  
  $\lim_{x \to -4^+} f(x) = 2$
- **Two-sided limit**:  
  $\lim_{x \to -4} f(x)$ **DNE**  
  *(Left and right limits disagree)*

### **(b) At $a = -1$**

- **Function value**:  
  $f(-1) = 4$
- **Left-hand limit**:  
  $\lim_{x \to -1^-} f(x) = 4$
- **Right-hand limit**:  
  $\lim_{x \to -1^+} f(x) = 4$
- **Two-sided limit**:  
  $\lim_{x \to -1} f(x) = 4$

### **(c) At $a = 2$**

- **Function value**:  
  $f(2) = -1$
- **Left-hand limit**:  
  $\lim_{x \to 2^-} f(x) = -1$
- **Right-hand limit**:  
  $\lim_{x \to 2^+} f(x) = 5$
- **Two-sided limit**:  
  $\lim_{x \to 2} f(x)$ **DNE**  
  *(Different left/right limits)*


### **(d) At $a = 4$**

- **Function value**:  
  $f(4)$ **DNE**  
  *(No defined point)*
- **Left-hand limit**:  
  $\lim_{x \to 4^-} f(x) = 2$
- **Right-hand limit**:  
  $\lim_{x \to 4^+} f(x) = 2$
- **Two-sided limit**:  
  $\lim_{x \to 4} f(x) = 2$
