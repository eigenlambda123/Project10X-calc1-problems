### **Problems**

Below is the graph of $f(x)$. For each of the given points determine:

1. The value of $f(a)$
2. $\lim_{x \to a^-} f(x)$ (left-hand limit)
3. $\lim_{x \to a^+} f(x)$ (right-hand limit)
4. $\lim_{x \to a} f(x)$ (two-sided limit)

If any of these quantities do not exist, clearly explain why.

#### Points to Evaluate:

- (a) $a = -2$
- (b) $a = 1$
- (c) $a = 3$
- (d) $a = 5$

### **Solutions**

### **(a) At $a = -2$**

- **Function value**:  
  $f(-2) = -1$  
  *(Defined point on graph)*
- **Left-hand limit**:  
  $\lim_{x \to -2^-} f(x)$ **DNE**  
  *(Function does not approach a finite value from the left)*
- **Right-hand limit**:  
  $\lim_{x \to -2^+} f(x) = -1$
- **Two-sided limit**:  
  $\lim_{x \to -2} f(x)$ **DNE**  
  *(Left limit DNE while right limit exists)*

### **(b) At $a = 1$**

- **Function value**:  
  $f(1) = 4$
- **Left-hand limit**:  
  $\lim_{x \to 1^-} f(x) = 3$
- **Right-hand limit**:  
  $\lim_{x \to 1^+} f(x) = 3$
- **Two-sided limit**:  
  $\lim_{x \to 1} f(x) = 3$


### **(c) At $a = 3$**

- **Function value**:  
  $f(3) = -2$
- **Left-hand limit**:  
  $\lim_{x \to 3^-} f(x) = 1$
- **Right-hand limit**:  
  $\lim_{x \to 3^+} f(x) = -3$
- **Two-sided limit**:  
  $\lim_{x \to 3} f(x)$ **DNE**  
  *(Left and right limits disagree)*


### **(d) At $a = 5$**

- **Function value**:  
  $f(5) = 4$
- **Left-hand limit**:  
  $\lim_{x \to 5^-} f(x) = 4$
- **Right-hand limit**:  
  $\lim_{x \to 5^+} f(x) = 4$
- **Two-sided limit**:  
  $\lim_{x \to 5} f(x) = 4$
