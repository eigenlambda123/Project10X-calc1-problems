```python
import matplotlib.pyplot as plt
import numpy as np

def plot_unit_circle_triangle(angle_deg, label):
    angle_rad = np.deg2rad(angle_deg)
    x = np.cos(angle_rad)
    y = np.sin(angle_rad)

    fig, ax = plt.subplots()
    ax.set_aspect('equal')

    # Draw unit circle
    circle = plt.Circle((0, 0), 1, fill=False)
    ax.add_artist(circle)

    # Terminal side
    ax.plot([0, x], [0, y], color='blue', label=f'{angle_deg}°')

    # Triangle sides
    ax.plot([x, x], [0, y], linestyle='--', color='gray')  # vertical leg
    ax.plot([0, x], [0, 0], linestyle='--', color='gray')  # horizontal leg

    # Point at (x, y)
    ax.scatter([x], [y], color='red')
    ax.text(x + 0.05, y + 0.05, f'({round(x, 2)}, {round(y, 2)})', fontsize=9)

    # Labels
    ax.text(0.1, 0.05, f'{label}', fontsize=10, color='green')

    # Axis & grid
    ax.set_xlim(-1.1, 1.1)
    ax.set_ylim(-1.1, 1.1)
    ax.axhline(0, color='gray', linewidth=0.5)
    ax.axvline(0, color='gray', linewidth=0.5)
    plt.title(f'Unit Circle: {angle_deg}°')
    plt.grid(True)
    plt.legend()
    plt.show()
```

# **Evaluate the following trigonometric expressions using special triangles**

Use the **special right triangles**:

- **30°-60°-90° triangle**:  
    Hypotenuse = 1  
    Short leg = $\dfrac{1}{2}$  
    Long leg = $\dfrac{\sqrt{3}}{2}$
    
- **45°-45°-90° triangle**:  
    Legs = $\dfrac{\sqrt{2}}{2}$  
    Hypotenuse = 1
    

---

**(a)** $\cos 30^\circ$  
$\quad= \dfrac{\text{adjacent}}{\text{hypotenuse}} = \dfrac{\sqrt{3}}{2}$


```python
plot_unit_circle_triangle(30, 'cos 30° = √3/2')
```

---
**(b)** $\sin 60^\circ$  
$\quad= \dfrac{\text{opposite}}{\text{hypotenuse}} = \dfrac{\sqrt{3}}{2}$

```python
plot_unit_circle_triangle(60, 'sin 60° = √3/2')
```

---
**(c)** $\csc 45^\circ$  
$\quad= \dfrac{1}{\sin 45^\circ} = \dfrac{1}{\frac{\sqrt{2}}{2}} = \sqrt{2}$

```python
plot_unit_circle_triangle(45, 'csc 45° = √2')
```

---
**(d)** $\cot 45^\circ$  
$\quad= \dfrac{1}{\tan 45^\circ} = 1$

```python
plot_unit_circle_triangle(45, 'cot 45° = 1')
```

---
**(e)** $\sin 30^\circ$  
$\quad= \dfrac{\text{opposite}}{\text{hypotenuse}} = \dfrac{1}{2}$

```python
plot_unit_circle_triangle(30, 'sin 30° = 1/2')
```

---
**(f)** $\sec 45^\circ$  
$\quad= \dfrac{1}{\cos 45^\circ} = \dfrac{1}{\frac{\sqrt{2}}{2}} = \sqrt{2}$

```python
plot_unit_circle_triangle(45, 'sec 45° = √2')
```

---
**(g)** $\tan 60^\circ$  
$\quad= \dfrac{\text{opposite}}{\text{adjacent}} = \dfrac{\sqrt{3}}{1} = \sqrt{3}$

```python
plot_unit_circle_triangle(60, 'tan 60° = √3')
```

---

**(h)** $\cos 45^\circ$  
$\quad= \dfrac{\text{adjacent}}{\text{hypotenuse}} = \dfrac{\sqrt{2}}{2}$

```python
plot_unit_circle_triangle(45, 'cos 45° = √2/2')
```
