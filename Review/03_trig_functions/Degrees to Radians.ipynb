# **Convert the following degree measures into radians**

Use the formula:

$$
\text{Radians} = \theta^\circ \times \frac{\pi}{180}
$$


---

### **Problems and Solutions**

**(a)** $45^\circ$  
$\quad= 45 \times \dfrac{\pi}{180} = \dfrac{\pi}{4}$


**(b)** $76^\circ$  
$\quad= 76 \times \dfrac{\pi}{180} = \dfrac{19\pi}{45}$


**(c)** $510^\circ$  
$\quad= 510 \times \dfrac{\pi}{180} = \dfrac{17\pi}{6}$


**(d)** $-240^\circ$  
$\quad= -240 \times \dfrac{\pi}{180} = \dfrac{-4\pi}{3}$


**(e)** $0^\circ$  
$\quad= 0 \times \dfrac{\pi}{180} = 0$


**(f)** $150^\circ$  
$\quad= 150 \times \dfrac{\pi}{180} = \dfrac{5\pi}{6}$


**(g)** $40^\circ$  
$\quad= 40 \times \dfrac{\pi}{180} = \dfrac{2\pi}{9}$


**(h)** $270^\circ$  
$\quad= 270 \times \dfrac{\pi}{180} = \dfrac{3\pi}{2}$


**(i)** $120^\circ$  
$\quad= 120 \times \dfrac{\pi}{180} = \dfrac{2\pi}{3}$


**(j)** $10^\circ$  
$\quad= 10 \times \dfrac{\pi}{180} = \dfrac{\pi}{18}$


**(k)** $50^\circ$  
$\quad= 50 \times \dfrac{\pi}{180} = \dfrac{5\pi}{18}$


**(l)** $-30^\circ$  
$\quad= -30 \times \dfrac{\pi}{180} = \dfrac{-\pi}{6}$


**(m)** $6^\circ$  
$\quad= 6 \times \dfrac{\pi}{180} = \dfrac{\pi}{30}$


**(n)** $300^\circ$  
$\quad= 300 \times \dfrac{\pi}{180} = \dfrac{5\pi}{3}$


**(o)** $-24^\circ$  
$\quad= -24 \times \dfrac{\pi}{180} = \dfrac{-2\pi}{15}$
