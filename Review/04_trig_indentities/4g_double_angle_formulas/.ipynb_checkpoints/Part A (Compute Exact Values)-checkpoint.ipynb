### **Problems**

**1.** Find $\sin(2\theta)$ if $\sin(\theta) = \dfrac{3}{5}$ and $\theta$ is in Quadrant I  
**2.** Find $\cos(2\theta)$ using $\cos(\theta) = \dfrac{4}{5}$ and $\theta$ in Quadrant I  
**3.** Find $\tan(2\theta)$ if $\tan(\theta) = \dfrac{2}{3}$  
**4.** If $\cos(\theta) = \dfrac{5}{13}$ and $\theta$ is in Quadrant I, find $\cos(2\theta)$ using all three forms  
**5.** Evaluate $\sin(2 \cdot 45^\circ)$ and $\cos(2 \cdot 30^\circ)$ exactly

---
### **Solutions**

**1.** Given: $\sin(\theta) = \dfrac{3}{5}$ and $\theta$ in Quadrant I  
We first find $\cos(\theta)$ using $\sin^2(\theta) + \cos^2(\theta) = 1$:

$$
\cos^2(\theta) = 1 - \left(\dfrac{3}{5}\right)^2 = 1 - \dfrac{9}{25} = \dfrac{16}{25}
\Rightarrow \cos(\theta) = \dfrac{4}{5}
$$  
(Since we're in Quadrant I, cosine is positive.)

Now use the identity:
$$
\sin(2\theta) = 2\sin(\theta)\cos(\theta) = 2 \cdot \dfrac{3}{5} \cdot \dfrac{4}{5} = \dfrac{24}{25}
$$

**Answer:**  
$\boxed{\dfrac{24}{25}}$


**2.** Given: $\cos(\theta) = \dfrac{4}{5}$, $\theta$ in Quadrant I  
We find $\sin(\theta)$:

$$
\sin^2(\theta) = 1 - \left(\dfrac{4}{5}\right)^2 = 1 - \dfrac{16}{25} = \dfrac{9}{25}
\Rightarrow \sin(\theta) = \dfrac{3}{5}
$$

Now apply the identity:
$$
\cos(2\theta) = \cos^2(\theta) - \sin^2(\theta)
= \left(\dfrac{4}{5}\right)^2 - \left(\dfrac{3}{5}\right)^2 = \dfrac{16 - 9}{25} = \dfrac{7}{25}
$$

**Answer:**  
$\boxed{\dfrac{7}{25}}$


**3.** Given: $\tan(\theta) = \dfrac{2}{3}$  
Use:
$$
\tan(2\theta) = \dfrac{2\tan(\theta)}{1 - \tan^2(\theta)} = \dfrac{2 \cdot \frac{2}{3}}{1 - \left(\frac{2}{3}\right)^2}
= \dfrac{\frac{4}{3}}{1 - \frac{4}{9}} = \dfrac{\frac{4}{3}}{\frac{5}{9}} = \dfrac{4}{3} \cdot \dfrac{9}{5} = \dfrac{12}{5}
$$

**Answer:**  
$\boxed{\dfrac{12}{5}}$


**4.** Given: $\cos(\theta) = \dfrac{5}{13}$, Quadrant I  
Then $\sin(\theta) = \sqrt{1 - \left(\dfrac{5}{13}\right)^2} = \sqrt{1 - \dfrac{25}{169}} = \sqrt{\dfrac{144}{169}} = \dfrac{12}{13}$

Now compute $\cos(2\theta)$ in all 3 forms:

- (a) $\cos^2(\theta) - \sin^2(\theta)$:
$$
= \left(\dfrac{5}{13}\right)^2 - \left(\dfrac{12}{13}\right)^2 = \dfrac{25 - 144}{169} = \dfrac{-119}{169}
$$

- (b) $2\cos^2(\theta) - 1$:
$$
= 2 \cdot \left(\dfrac{5}{13}\right)^2 - 1 = 2 \cdot \dfrac{25}{169} - 1 = \dfrac{50}{169} - \dfrac{169}{169} = \dfrac{-119}{169}
$$

- (c) $1 - 2\sin^2(\theta)$:
$$
= 1 - 2 \cdot \left(\dfrac{12}{13}\right)^2 = 1 - 2 \cdot \dfrac{144}{169} = 1 - \dfrac{288}{169} = \dfrac{-119}{169}
$$

**Answer:**  
$\boxed{\cos(2\theta) = -\dfrac{119}{169}}$ (by all 3 forms)


**5.** Evaluate:
- $\sin(2 \cdot 45^\circ) = \sin(90^\circ) = \boxed{1}$  
- $\cos(2 \cdot 30^\circ) = \cos(60^\circ) = \boxed{\dfrac{1}{2}}$
